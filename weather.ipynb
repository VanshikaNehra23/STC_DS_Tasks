##  CLEANING THE DATA

```python
df_2=df.fillna(method='bfill',axis=0).fillna(0)
```

```python
df_2.info()
```

```python
df_2.head()
```

```python
def converter(x):
    if x=='Yes':
        return 1
    else:
        return 0
```

```python
data=df_2.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm'],axis=1)
data['RainTod']=data['RainToday'].apply(converter)
data['RainTom']=data['RainTomorrow'].apply(converter)
data.drop(['RainToday','RainTomorrow'],axis=1,inplace=True)
data
```

##  DECISION TREE CLASSIFIER

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('RainTom',axis=1)
y=data['RainTom']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
pred=dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

##  RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF=RandomForestClassifier()
```

```python
RF.fit(X_train,y_train)
```

```python
pred=RF.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

##  SUPPORT VECTOR MACHINE

```python
pred=svc.predict(X_test)
```

##  GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1,1],'gamma':[1,0.1]}
```

```python
gs=GridSearchCV(SVC(),param_grid,verbose=2)
gs.fit(X_train,y_train)
```

```python
pred=gs.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
